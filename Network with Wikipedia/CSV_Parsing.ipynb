```python
import pandas as pd
```

```python
df = pd.read_csv('/content/original nodes.csv')
df.head()
```

```python
df_300 = df[df.Degree >= 300]
print(f"{df_300.shape[0]} nodes")
df_300
```

```python
list_group = []
for i in df_300.itertuples():
  if i.Degree <= 499:
    list_group.append('Group 5')
  elif i.Degree <= 699:
    list_group.append('Group 4')
  elif i.Degree <= 999:
    list_group.append('Group 3')
  elif i.Degree <= 1199:
    list_group.append('Group 2')
  else:
    list_group.append('Group 1')

list_group[0:10]
```

```python
df_300['degree_group'] = list_group
df_300.head()
```

```python
df_300.to_csv('gephi-groups-300.csv', index=False)
```
